```python
import pandas as pd
```

```python
#Dataframe from dictionary
waste_data = {"waste_type": ["Plastic", "Organic", "Electronic"],
              "quantity": [12, 15, 30],
              "recyclable": ["Yes", "NO", "Yes"]
             }
waste_data_df = pd.DataFrame(waste_data)
waste_data_df
```

```python
#Dataframe using list
waste_list = [["Plastic", 15, True], ["Organic", 25, False],
             ["Electronic", 35, True]]
waste_list_df = pd.DataFrame(waste_list, columns = ["Waste_type", "Quantity", "Recyclable"])
waste_list_df
```

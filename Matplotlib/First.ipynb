```python
import matplotlib.pyplot as plt
```

```python
x = [2,4,6,8,10]
y = [30,20,60,40,50]
plt.figure(figsize = (3,3))
plt.plot(x,y,marker='*' ,mfc = 'g',mec='k',color = 'r',linestyle = '--',ms=10)
plt.title("This is my first plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid()
plt.show()
```

```python
import numpy as np
x = np.linspace(10,100,1000)
y = np.sin(x)
plt.figure(figsize = (3,3))
plt.plot(x,y)
plt.title("Sine Wave")
plt.xlabel("X vlaue")
plt.ylabel("sin(x)")
plt.show()
```
